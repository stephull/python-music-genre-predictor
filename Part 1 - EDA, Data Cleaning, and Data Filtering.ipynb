```python
#
# convert genres to array
#

def delimit_genres(li):
    ret_arr = []
    for line in li:
        ret_arr.append(str(line).split(';'))
    return ret_arr

artists['Genres'] = delimit_genres(artists['Genres'])
print(artists.shape)
artists.head()
```

```python
#
# Only get English lyrics
#

lyrics = lyrics[lyrics['language'] == 'en']
lyrics.reset_index(drop=True, inplace=True)
print(lyrics.shape)
lyrics.head()
```

```python
#
# filter out any artists that don't contain English songs
# (takes some time)
#

for i in range(len(artists)):
    link = artists['Link'][i]
    lyrics_temp = lyrics[lyrics['ALink'] == link]
    temptable = lyrics_temp[lyrics_temp['language'] == 'en']
    if temptable.empty:
        artists.drop(artists.loc[artists['Link'] == link].index, inplace=True)
artists.reset_index(drop=True, inplace=True)
lyrics.drop(['language'], axis=1, inplace=True)
artists
```

```python
#
# Filter out unneeded or irrelevant genres
#
replaceable = [
    'Romântico', 'Gospel/Religioso', 'Trilha Sonora', 'Rock Alternativo', 'Clássico', 
    'Progressivo', 'Tecnopop', 'Gótico', 'Infantil', 'Psicodelia',
    'Soul Music', 'Hard Rock', 'Punk Rock', 'Soft Rock', 'Trip-hop'
    
]
replacewith = [
    'Romantic' , "Gospel", 'Soundtack', "Rock", 'Classic',
    "Progressive Music", 'Pop', 'Rock', "Children's music", "Psychedelic music",
    'Soul', 'Rock', 'Rock', 'Rock', 'Hip Hop'
]

# all genres with 5 or less songs, and presumed to be majority non-English
deleteable = [
    "Axé", "K-Pop/K-Rock", "J-Pop/J-Rock", "MPB", "Sertanejo", "Bossa Nova", "Brazillian Punk",
    "Samba", "COLETÂNEA", "Jovem Guarda", "Pagode", "Velha Guarda", "Fado", "Kizomba", "Regional", "Electro Swing",
    "Trap", "Forró", 'Funk Carioca', 'Pós-Punk', 'New Age'
]

# similar to list above, but only take out specified genre, not whole row of data
semi_deleteable = [
    "Black Music", " Black Music ", "World Music", " World Music ", "Piano Rock", "Post-punk", "Power-Pop", "nan"
]

for i in range(len(artists['Genres'])):
    ltemp = artists['Genres'][i]
    for j in range(len(ltemp)):
        gen = ltemp[j].strip()
        # check for deleteable words
        if gen in deleteable:
            artists.drop([i], inplace=True)
            break
        # check for replaceable words
        if gen in replaceable:
            temp = replaceable.index(gen)
            ltemp[j] = str(replacewith[temp])

for i in range(len(artists['Genres'])):
    ltemp = artists['Genres'].iloc[i]
    j = 0
    while j < len(ltemp):
        # check for redundant genres
        gen = ltemp[j].lstrip()
        if gen in semi_deleteable or gen.strip() in semi_deleteable:
            ltemp.remove(ltemp[j])
            artists['Genres'].iloc[i] = ltemp
        j += 1
artists.reset_index(drop=True, inplace=True)
artists
```

```python
# one more time, clear all empty rows
for i in range(artists.shape[0]):
    genres_list = artists['Genres'].loc[i]
    if len(genres_list) == 0:
        artists.drop([i], inplace=True)
artists.reset_index(drop=True, inplace=True)
artists
```

```python
# just in case, any genres that are slashed gets randomized by chance
for i in range(artists.shape[0]):
    for j in range(len(artists['Genres'].loc[i])):
        if "/" in artists['Genres'].loc[i][j]:
            lit = artists['Genres'].loc[i][j].split("/")
            randomized = np.random.choice([0,1])
            artists['Genres'].loc[i][j] = lit[0] if randomized == 0 else lit[1]
```

```python
# last but not least, take out the songs and popularity columns
artists.drop(['Songs', 'Popularity'], axis=1, inplace=True)
lyrics.drop(['SName', 'Lyric'], axis=1, inplace=True)
```

```python
# SOURCE: https://www.delftstack.com/howto/python/remove-numbers-from-string-python/
import bs4 as bs
import urllib.request
import string
from string import digits
from urllib.error import HTTPError

import nltk
nltk.download('stopwords')
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
nltk.download('punkt')

'''
    FUNCTIONS
'''
# functions to grab URL link
def refine_lyrics(url, p):
    if p: print(f"Looking for {url} ...\n")
    try: 
        source = urllib.request.urlopen(url).read()
    except HTTPError:
        if p: 
            print("URL not found, try again.")
    try: 
        soup = bs.BeautifulSoup(source, 'lxml')
    except:
        if p: 
            print("Unable to recall source.\n")
    request = soup.find('div', {"id": 'lyrics'})
    text = request.get_text(separator=' ')
    return text

def defluff_text(src):
    temp = str(src).replace(' ', '')
    temp = temp.translate(str.maketrans("/(')\p{P}/", "/(')\p{P}/", digits)).partition('\n')[0]
    return temp 

def request_lyrics(row, prints=False):
    link = defluff_text(row['SLink'])
    url = f"{LYRIC_URL}{link}"
    ret = refine_lyrics(url, prints)
    
    almost_there = ret.split(" ")
    almost_there = [str(word).lower() for word in almost_there]
    almost_there = "".join(a+" " for a in almost_there)
    
    ret = almost_there
    return ret
```

```python
# count each word in the song
try:
    import enchant
except ModuleNotFoundError:
    !{sys.executable} -m pip install pyenchant
    import enchant
    
denchant = enchant.Dict("en_US")

def stopStopWords(d):
    new_d = {**d}
    for k in new_d.keys():
        if k in stops:
            del new_d[k]
    return {**new_d}

def countEachWord(source):
    words = source.split(' ')
    rset = {}
    
    words = [str(word).lower().strip() for word in words]
    
    for i in range(len(words)):        
        #https://datagy.io/python-remove-punctuation-from-string/
        words[i] = words[i].translate(str.maketrans("/(')\p{P}/", "/(')\p{P}/", string.punctuation))
        
        if len(words[i]) == 0:
            continue
        if denchant.check(words[i]) == False:
            continue
        if words[i] in stops:
            continue
        
        # put to word set, count how many times a word is used.
        rset.update({words[i]:rset.get(words[i])+1} if words[i] in rset else {words[i]:1})
    return rset

wordset = countEachWord(returned_string)
```

```python
import time
from datetime import datetime

GW = "GenreWords"  # folder
GWDICT = {}  # dictionary

# get dictionary for all genre words and 
# store all data into multiple .csv files using 'GenreWords':
def tabulateDict(genres, lyrics):
    assert type(genres) == list and type(lyrics) == dict
    for g in genres:
        g = g.strip()
        
        # introduce new genre if unavailable
        if g not in GWDICT.keys():
            GWDICT[g] = lyrics
            continue
            
        # existing genre, make deep copy of dictionary
        newset = {**GWDICT[g]}
        for l in lyrics.keys():
            newset[l] = lyrics[l] if l not in newset else newset[l] + lyrics[l]
        GWDICT[g] = {**newset}
```
